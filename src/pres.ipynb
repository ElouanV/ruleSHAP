```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Generate a graph with a center node and 3 neighbors
G = nx.Graph()
G.add_node("center")
G.add_nodes_from(["a", "b", "c"])
G.add_edges_from([("center", "a"), ("center", "b"), ("center", "c")])
# Create 6 new nodes, each connected to one of the 3 neighbors
G.add_nodes_from(["d", "e", "f", "g", "h", "i"])
G.add_edges_from([("a", "d"), ("a", "e"), ("a", "f"), ("b", "g"), ("b", "h"), ("b", "i")])


# Add 6 more nodes connected to  d e f g h i
G.add_nodes_from(["j", "k", "l", "m", "n", "o"])
G.add_edges_from([("d", "j"), ("d", "k"), ("d", "l"), ("e", "m"), ("e", "n"), ("e", "o")])

# Draw the graph
nx.draw(G, with_labels=True, node_color="pink", node_size=1500, edge_color="green")
```

```python
# Take node a as center
random_ = "a"

# Show the neighbors of the random node with a distance of 3
subgraph = nx.ego_graph(G, random_, radius=3)
layout = nx.spring_layout(subgraph)
for i in range(1,4):
    # Draw the subgraph with nodes further than i from the random node with transparent color
    nx.draw_networkx_nodes(subgraph, layout, nodelist=[n for n in subgraph.nodes() if nx.shortest_path_length(subgraph, random_, n) > i], node_size=100, alpha=0.2, node_color='b')
    # Draw the subgraph with nodes closer than i from the random node with opaque color
    nx.draw_networkx_nodes(subgraph, layout, nodelist=[n for n in subgraph.nodes() if nx.shortest_path_length(subgraph, random_, n) <= i], node_size=100, node_color='g')
    # Draw the subgraph with edges further than i from the random node with transparent color
    nx.draw_networkx_nodes(subgraph, layout, nodelist=[random_],node_color='r', node_size=200)

    nx.draw_networkx_edges(subgraph, layout)
    plt.show()
```
